# Квантовые эмбэддинги и Feature Map

При построении бинарного классификатора с помощью туториала [Neural Network Classifier & Regressor](https://qiskit.org/documentation/machine-learning/tutorials/02_neural_network_classifier_and_regressor.html) мы использовали [ZZFeatureMap](https://qiskit.org/documentation/stubs/qiskit.circuit.library.ZZFeatureMap.html) карту признаков, позволяющая представить входные данные в виде понятно для работы с квантовым классификатором. Попробуем разобраться, как работают карты признаков

В глоссарии [pennelane](https://pennylane.ai/qml/glossary.html) приводится определение [feature map](https://pennylane.ai/qml/glossary/quantum_feature_map.html). Feature map может рассматриваться как функция сопоставления множества веторов в исходнов пространстве $X$ множеству векторов, но уже в некотором пространстве пространстве признаков $F$.

При этом в случае QML пространство признаков $F$ обладает следующими свойствами:
1. Является [гильбертовым](https://semyonsinchenko.github.io/qmlcourse/_build/html/book/linalgblock/bra-ket.html#id2)
1. Векторы в пронстве $F$ являются квантовыми состояниями, то есть $x \rightarrow |\phi(x)\rangle$

Трансформация в квантовое состояние происходит с помощью применения [унитарного преобразования](https://semyonsinchenko.github.io/qmlcourse/_build/html/book/linalgblock/bra-ket.html#id6) $U_{\phi}(x)$, коим может выступать [variational circuit](https://pennylane.ai/qml/glossary/variational_circuit.html)

Полученный вектор называют [квантовым эмбддингом](https://pennylane.ai/qml/glossary/quantum_embedding.html)

## [Basis Embedding](https://pennylane.ai/qml/glossary/quantum_embedding.html#basis-embedding)

Самым простым способом получения квантовых эмбиддингов является сопоставления классических битов к кубитам, при условии, что исходные данные могут быть представлены битовой строкой. 

Пример представления можно найти в [глоссарии](https://pennylane.ai/qml/glossary/quantum_embedding.html#basis-embedding)

Минусы такого подхода очевидны - нужно огромное число кубитов, чтобы закодировать какие-то осмысленные данные

### [Amplitude Embedding](https://pennylane.ai/qml/glossary/quantum_embedding.html#amplitude-embedding)

Другой вариант - закодировать вектор как амплитуду кубита: пример можно найти в [глоссарии](https://pennylane.ai/qml/glossary/quantum_embedding.html#amplitude-embedding)
